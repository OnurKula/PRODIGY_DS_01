```python
pplt_1960 = data.loc["1960"].dropna()
plt.figure(figsize=(10, 6))
plt.hist(pplt_1960, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Population Across Countries in 1960')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```

```python
pplt_2023 = data.loc["2023"].dropna()
plt.figure(figsize=(10, 6))
plt.hist(pplt_2023, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Population Across Countries in 2023')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

Turkiye = data.loc["1960":"2023", "Turkiye"]
Turkiye.index = Turkiye.index.astype(int)


ax.plot(Turkiye.index,Turkiye.values)
ax.set_xlabel("Year")
ax.set_ylabel("Population")
plt.title('Turkey Population Growth (1960-2023)', fontsize=16)
# Set x-ticks at intervals of 5 years to avoid cluttering
ax.set_xticks(range(1960, 2024, 5))

ax.set_xticklabels(range(1960, 2024, 5), rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Extract population data for Turkey, India, China, and Peru
countries = ["Turkiye", "India", "China", "Peru"]
population_data = data.loc["1960":"2023", countries]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot for Turkey
axs[0, 0].plot(population_data.index.astype(int), population_data["Turkiye"], color='skyblue', marker='o')
axs[0, 0].set_title('Turkey Population (1960-2023)')
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('Population')
axs[0,0].set_xticks(range(1960, 2024, 5))

axs[0,0].set_xticklabels(range(1960, 2024, 5), rotation=45)

# Plot for India
axs[0, 1].plot(population_data.index.astype(int), population_data["India"], color='green', marker='o')
axs[0, 1].set_title('India Population (1960-2023)')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('Population')
axs[0,1].set_xticks(range(1960, 2024, 5))

axs[0,1].set_xticklabels(range(1960, 2024, 5), rotation=45)

# Plot for China
axs[1, 0].plot(population_data.index.astype(int), population_data["China"], color='red', marker='o')
axs[1, 0].set_title('China Population (1960-2023)')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Population')
axs[1,0].set_xticks(range(1960, 2024, 5))

axs[1,0].set_xticklabels(range(1960, 2024, 5), rotation=45)

# Plot for Peru
axs[1, 1].plot(population_data.index.astype(int), population_data["Peru"], color='purple', marker='o')
axs[1, 1].set_title('Peru Population (1960-2023)')
axs[1, 1].set_xlabel('Year')
axs[1, 1].set_ylabel('Population')
axs[1,1].set_xticks(range(1960, 2024, 5))

axs[1,1].set_xticklabels(range(1960, 2024, 5), rotation=45)

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
top_10_2023 = data.loc["2023"].sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_10_2023.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Populated Countries in 2023')
plt.ylabel('Population')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
country = "India"
plt.figure(figsize=(10, 6))
plt.plot(data.index.astype(int), data[country], marker='o', color='green')
plt.title(f'{country} Population Growth (1960-2023)')
plt.ylabel('Population')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Choose countries to compare
countries = ["China", "India", "United States", "Brazil"]

# Plot population growth for selected countries
plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot(data.index.astype(int), data[country], marker='o', label=country)

plt.title('Population Growth Comparison (1960-2023)')
plt.ylabel('Population')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

selected_countries = ["China", "India", "United States", "Brazil", "Indonesia"]
heatmap_data = data[selected_countries].T  # Transpose for heatmap

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=False)
plt.title('Population Heatmap (1960-2023)')
plt.ylabel('Country')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
growth_rate = data["China"].pct_change() * 100

plt.figure(figsize=(10, 6))
plt.plot(data.index.astype(int), growth_rate, marker='o', color='red')
plt.title('China Population Growth Rate (1960-2023)')
plt.ylabel('Growth Rate (%)')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
growth_rate = data["India"].pct_change() * 100

plt.figure(figsize=(10, 6))
plt.plot(data.index.astype(int), growth_rate, marker='o', color='red')
plt.title('India Population Growth Rate (1960-2023)')
plt.ylabel('Growth Rate (%)')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```
